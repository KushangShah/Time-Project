1. Create a class Time which takes two inputs: hours and minutes.
2. Instantiate the values using a function.
3. Then, define two functions for displaying the time in the format explained in the question and for displaying the ratio of hours to minutes.
4. Use try-except block for displaying the ratio to account for Zero Division Error.

```python
# creating a class time
class Time:

  def __init__(self, hour, minute):
    # assigning parameter for hour and minute
    self.hour = hour
    self.minute = minute

    # Defining a Display time
  def DisplayTime(self):

    try:
      # In clock hour are limited only up to 24 hours
      if self.hour > 23:
        print("Hour should not be more than 24")
        return

      # Minites are limited to 60 as well
      if self.minute > 59:
        print("Minute should not be more than 60")
        return

      # If time is afternoon [PM]
      if self.hour >= 12:
        am_pm = "PM"
        if self.hour > 12:
          self.hour -= 12

      # If it's morning [AM]
      else:
        am_pm = "AM"
        if self.hour == 0:
          self.hour = 12

      #printing the time
      print(f"The Time is {self.hour}:{self.minute} {am_pm}")

    except Exception as e:
      print(f"An error occured: {e}")


      # Defining a Display ratio time with zero divisional error
  def DisplayRatio(self):
    try:
      ratio = self.minute / self.hour
    except ZeroDivisionError:
      print("An error occured: Cannot divide by 0")
```

## OUTPUT EXAMPLE

```python
# a list of hour, minut
hour_min_list = [(23,45), (34,50), (12,34), (14,67),(19,20), (2,15), (0, 10)]

# Assignig for loop to iterate over each (hour, minut)
for hour, minute in hour_min_list:
  t = Time(hour, minute)
  t.DisplayTime()
  t.DisplayRatio()
  print()
```
